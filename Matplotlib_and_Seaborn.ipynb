### *formatting line plot*

#### *- title and label*

```python
# provide title and x-y label

plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('KOHLI vs SHARMA IPL CARRER')
plt.xlabel('SEASON')
plt.ylabel('RUNS SCORED')
```

#### *- color*

```python
# what if want to change the colour

plt.plot(batsman['index'],batsman['V Kohli'],color='red')
plt.plot(batsman['index'],batsman['RG Sharma'],color ='#0A64F5')# <-- can provide any hex code

plt.title('KOHLI vs SHARMA IPL CARRER')
plt.xlabel('SEASON')
plt.ylabel('RUNS SCORED')
```

#### *- linestyle and linewidth*

```python
# changing the linestyle and linewidth

# linestyle -> solid(d)/dashed/dotted/dashdot or -/--/:/-.
# linewidth -> 1.5(d)

plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle = 'dashed',linewidth=2)
plt.plot(batsman['index'],batsman['RG Sharma'],color ='#0A64F5',linestyle ='-.',linewidth = 1)

plt.title('KOHLI vs SHARMA IPL CARRER')
plt.xlabel('SEASON')
plt.ylabel('RUNS SCORED')
```

#### *- marker and markersize*

```python
# manipulate markers and markersize
'''
'o': circle
's': square
'*': star
'+': plus sign
'x': x mark
'^': triangle up
'v': triangle down
'D': diamond
'.': point (small dot)
',': pixel (tiny dot)
'h': hexagon
'p': pentagon
 '''
 # markersize or ms  -> default is 6

plt.plot(batsman['index'],batsman['V Kohli'],color='red',marker='x',markersize = 15)
plt.plot(batsman['index'],batsman['RG Sharma'],color ='#0A64F5',marker ='o')

plt.title('KOHLI vs SHARMA IPL CARRER')
plt.xlabel('SEASON')
plt.ylabel('RUNS SCORED')
```

#### *- labels and legend*

```python
# adding lables and legends
# -> we have to provide lables first to see the legends

plt.plot(batsman['index'],batsman['V Kohli'],color='red',marker='x',label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color ='#0A64F5',marker ='o',label ='Rohit')

plt.title('KOHLI vs SHARMA IPL CARRER')
plt.xlabel('SEASON')
plt.ylabel('RUNS SCORED')
plt.legend(loc= 'best') # -> all options exist
```

## *# Subplots*

```python
# plotting multiple graphs
#-> alternate way of plotting graphs in plt
```

```python
plt.figure(figsize=(10,6))


plt.scatter(bat['avg'],bat['strike_rate'])

plt.title('Avg vs Strike Rate')
plt.xlabel('Avg')
plt.ylabel('Strike Rate')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,6)) # will generate figure obj-> graph and axis obj

ax.scatter(bat['avg'],bat['strike_rate'])

ax.set_title('Something')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike_rate')
fig.show()
```

```python
# can plot multiple graphs using this syntex

plt.subplots(nrows=2,ncols=2)
# defines the column and rows we want
# -> returns 4 axis object and can index them like matrix -> as it is 2D
```

```python
fig, ax  = plt.subplots(nrows=2,ncols=1,figsize=(10,6))
# ax -> will store array of axes and we can index them
ax[0].scatter(bat['avg'],bat['strike_rate'])  # <-- this will be indexed
ax[0].set_title('Average vs Strike Rate')
ax[0].set_xlabel('Average')
ax[0].set_ylabel('Strike_rate')


ax[1].scatter(bat['avg'],bat['runs'])  # <-- this will be indexed for second graph
ax[1].set_title('Average vs Runs')
ax[1].set_xlabel('Average')
ax[1].set_ylabel('runs')
```

```python
# add padding or gap between them -

fig, ax  = plt.subplots(nrows=2,ncols=1,figsize=(10,6))
ax[0].scatter(bat['avg'],bat['strike_rate'])
ax[0].set_title('Average vs Strike Rate')
ax[0].set_xlabel('Average')
ax[0].set_ylabel('Strike_rate')
#----------------------------------------------
ax[1].scatter(bat['avg'],bat['runs'])
ax[1].set_title('Average vs Runs')
ax[1].set_xlabel('Average')
ax[1].set_ylabel('runs')
#=========================================================

plt.tight_layout() # <-h_pad=3.0 this will add padding
# plt.subplots_adjust(hspace=0.4)
```

```python
# what if both graphs share same x axis -> and want same coordinates

fig, ax  = plt.subplots(nrows=2,ncols=1,figsize=(10,6),sharex=True)
#=========================================================
ax[0].scatter(bat['avg'],bat['strike_rate'],c='red')
ax[0].set_title('Average vs Strike Rate')
ax[0].set_xlabel('Average')
ax[0].set_ylabel('Strike_rate')
#----------------------------------------------
ax[1].scatter(bat['avg'],bat['runs'])
ax[1].set_title('Average vs Runs')
ax[1].set_xlabel('Average')
ax[1].set_ylabel('runs')
#---------------------------
plt.tight_layout()
```

```python
# plotting 4*4 graph

fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,6))

# here axis is 2D array and we need to index according to it

ax[0,0].scatter(bat['avg'],bat['strike_rate'])
#-----------------------------------------
ax[0,1].scatter(bat['avg'],bat['runs'])
#-----------------------------------------
ax[1,0].hist(bat['avg'])
#-----------------------------------------
ax[1,1].hist(bat['strike_rate'])
```

```python
# design 4*4 graph

fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,6))

# here axis is 2D array and we need to index according to it

ax[0,0].scatter(bat['avg'],bat['strike_rate'],color='red')
ax[0,0].set_title('Average vs Strike_rate')
ax[0,0].set_xlabel('Average')
ax[0,0].set_ylabel('Strike_rate')
#-----------------------------------------
ax[0,1].scatter(bat['avg'],bat['runs'],color='red')
ax[0,1].set_title('Average vs Runs')
ax[0,1].set_xlabel('Average')
ax[0,1].set_ylabel('Runs')
#-----------------------------------------
ax[1,0].hist(bat['avg'],color='black')
ax[1,0].set_title('Histogram_Average')
ax[1,0].set_xlabel('Average')
ax[1,0].set_ylabel('Total Number of Players')
#-----------------------------------------
ax[1,1].hist(bat['strike_rate'],color='black')
ax[1,1].set_title('Histogram_Strike_rate')
ax[1,1].set_xlabel('Strike_rate')
ax[1,1].set_ylabel('Total Number of Players')
#---------------------------------------------

plt.tight_layout()
fig.show()
```

```python
# basic logic is -> if we put 2 graphs together -> they will overlap one on top of other
#  -> unless we spacigy it via subplots
```

```python
# alternative syntax

fig = plt.figure()

ax1 = fig.add_subplot(2,2,1) # rows,column,position
ax1.scatter(bat['avg'],bat['strike_rate'],color='red')
#-----------------------------------------------------
ax2 = fig.add_subplot(2,2,2) # rows,column,position
ax2.hist(bat['strike_rate'],color='black')
#-----------------------------------------------------
ax3 = fig.add_subplot(2,2,(3,4)) # <--- for occupying both positions
ax3.scatter(bat['avg'],bat['runs'],color='black')
```

```python
# in previos syntax

# Define layout: 'A' and 'B' on top, 'C' spanning bottom
fig, ax = plt.subplot_mosaic([['A', 'B'],
                               ['C', 'C']],
                              figsize=(10, 6))

# Now ax is a dictionary with keys 'A', 'B', 'C'
ax['A'].scatter(bat['avg'], bat['strike_rate'])
ax['A'].set_title('Average vs Strike Rate')

ax['B'].scatter(bat['avg'], bat['runs'])
ax['B'].set_title('Average vs Runs')

ax['C'].hist(bat['avg'])
ax['C'].set_title('Average Distribution')

plt.tight_layout()
plt.show()

```

```python
# to plot loss function ->
```

```python
# x**2 + y**2

x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)
z = xx**2 + yy**2
```

```python
fig = plt.figure()

ax = fig.add_subplot(projection='3d')
k = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(k) # for colourbar -> we need to store it in variable to show
plt.show()
```

```python
z = np.sin(xx) + np.cos(yy)
```

```python
fig = plt.figure()

ax = fig.add_subplot(projection='3d')
k = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(k) # for colourbar -> we need to store it in variable to show
plt.show()
```

```python
# contour plot

# represents 3D graph to 2D - Upper graph to coutour -> seeing from top
```

```python
fig = plt.figure()

ax = fig.add_subplot()
k = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(k) # for colourbar -> we need to store it in variable to show
plt.show()
```

```python
# filled contour graph =>
fig = plt.figure()

ax = fig.add_subplot()
k = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(k) # for colourbar -> we need to store it in variable to show
```

```python
# Filled contour graph
fig = plt.figure(figsize=(12,5))

# Left plot - filled contour
ax1 = fig.add_subplot(1, 2, 1)
k = ax1.contourf(xx, yy, z, cmap='viridis')  # Changed ax to ax1
fig.colorbar(k, ax=ax1)  # Added ax parameter
ax1.set_title('Filled Contour')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')

# Right plot - line contour
ax2 = fig.add_subplot(1, 2, 2)
l = ax2.contour(xx, yy, z, cmap='viridis')  # Changed ax to ax2
fig.colorbar(l, ax=ax2)  # Added ax parameter
ax2.set_title('Line Contour')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')


plt.tight_layout()
plt.show()
```
